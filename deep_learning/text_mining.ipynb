```python
import tensorflow as tf
tf.keras.preprocessing.text.text_to_word_sequence('i am a boy')
tf.keras.preprocessing.text.text_to_word_sequence('it`s desk')

doc = ['먼저 텍스트의 각 단어를 나누어 토큰화 합니다.', '텍스트의 단어로 토큰화해야 딥러닝에서 인식됩니다.', '토큰화한 결과는 딥러닝에서 사용할 수 있습니다.']
token = tf.keras.preprocessing.text.Tokenizer()
token.fit_on_texts(doc)
token.word_docs
token.word_index
# tf.keras.utils.to_categorical(token.word_index)
token.texts_to_sequences(doc) # import tensorflow as tf
tf.keras.preprocessing.text.text_to_word_sequence('i am a boy')
tf.keras.preprocessing.text.text_to_word_sequence('it`s desk')

doc = ['먼저 텍스트의 각 단어를 나누어 토큰화 합니다.', '텍스트의 단어로 토큰화해야 딥러닝에서 인식됩니다.', '토큰화한 결과는 딥러닝에서 사용할 수 있습니다.']
token = tf.keras.preprocessing.text.Tokenizer()
token.fit_on_texts(doc)
token.word_docs
token.word_index
# tf.keras.utils.to_categorical(token.word_index)
token.texts_to_sequences(doc) # [[3, 1, 4, 5, 6, 7, 8], [1, 9, 10, 2, 11], [12, 13, 2, 14, 15, 16]]
tf.keras.utils.to_categorical(token.texts_to_sequences([doc[0]]), num_classes= len(token.texts_to_sequences(doc[0])) + 1) # one hot 인코딩 
# tf.keras.layers.Embedding # 레이어라 모델 시퀸스에 추가해서 씀
```

```python
from transformers import BertTokenizer, TFBertModel
import tensorflow as tf
# 토크나이저 및 모델 로드
tokenizer = BertTokenizer.from_pretrained('bert-large-uncased')
model = TFBertModel.from_pretrained('bert-large-uncased')

# 입력 문장 토큰화
text = "Hello, how are you?"
#inputs = tokenizer(text, return_tensors="pt")
inputs = tokenizer(text, return_tensors="tf")

# 모델을 통해 인코딩된 표현 얻기
outputs = model(**inputs)
print(outputs)
```
