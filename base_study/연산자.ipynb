```python
# 산술연산자 (사칙연사)
print('2 + 1 -> ', 2 + 1)
print('2 - 1 -> ', 2 - 1)
print('2 * 1 -> ', 2 * 1)
print('2 / 1 -> ', 2 / 1)
print('2 // 1 -> ', 2 // 1)
print('2 % 1 -> ', 2 % 1)
print('2 ** 1 -> ', 2 ** 1)
print('2 & 1 -> ', 2 & 1)
print('2 | 1 -> ', 2 | 1)
print('~1 -> ', ~1)  # 비트 반전 0000 0001 -> 1111 1110  음수 읽는 법은 반전후 + 1 (2의 보수)
print(bin(-2))
print('2 ^ 1 -> ', 2 ^ 1) # xor  A XOR B=(A and not B) or (not A and B) 두 개의 조건 중 하나만 참(True)일 때 참을 반환하는 논리 연산
```

```python
# 연산자 우선순위
# 괄호 먼저 젤 안쪽거
# 곱셈 나눗셈 먼저 앞쪽 부터
# 앞쪽 부터 계산
5 * (160 / 4 - (4 + 2) * 5 ) - 15
```

```python
# 복합 대입 연산자
# +=, -=, *=, /=, //=, %=, **=
num = 1
num += 1
print(num)
num *= 2
print(num)
print(str(5), int('5'), float(1))
```

```python
print(ord('1'))
print(bin(49))
print(ascii(0b110001))
# 문자열을 UTF-8 인코딩하여 바이트로 변환
byte_data = bytes("안녕", 'utf-8')

# 바이트 객체 출력
print(byte_data)  # \xec\x95\x88은 안 문자의 UTF-8 인코딩입니다. 한글은 3바이트 아스키는1바이트

```

```python
# 논리연산자 and or not
print(bool(0)) 
print(bool(-1), bool(1.0), bool('a'), bool(''), bool(None), bool({}), []) # 0을 제외한 모든게 참

# 연산자 operator  피연산자 operand

import time
time.sleep(1)
print('1초')
```
