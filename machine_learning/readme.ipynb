# 머신러닝 machine learning
- 인공지능 연구과제 중 하나
- 인간의 뇌가 자연스럽게 수행하는 '학습' 이라는 능력을 컴퓨터로  구현하는 방법
    1. 대규모 샘플 데이터 분석
    2. 규칙을 찾아냄 
    3. 규칙을 기반으로 다른 데이터 분류 및 예측 (인공지능 = 분류 및 예측)

- 벡터 vector : 공간에서 크기와 방향을 가진다
- 특징량 : 어떤 요소가 모여있는 것

## 예상 시나리오
0. 데이터 수집 
1. 데이터를 좌표에 찍는다 (데이터 가공)
2. x 최소 최대 y 최소 최대 를 기준으로 이미지 크기를 결정한다
3. 이미지의 크기를 축소하고 패턴크기로 정형화 한다 (비지도학습)
4. 이미지를 분석하여 어떤 군집을 이루는지 가려낸다 
5. 가려낸 패턴을 토대로 데이터를 정의 한다. (지도학습)

## 이미지 분석 시나리오
1. 특징을 찾아낸다.
2. 특징을 가려낸다.
3. 특징을 비교한다. 
4. 특징이 맞는지 확정한다.
