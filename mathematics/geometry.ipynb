```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
print(math.e) # 오일러 상수

p = 3 # 평행이동. 지수와 b에 값을 변화시켜도 그래프의 모양은 같다.
x = [i / 10 + p for i in range(-10,11)]
a = [2 * -1 for i in range(-10,11)] # 대칭이동. 지수나 밑에 -1 를 곱하면 대칭이 된다
y = [j ** i for i, j in zip(x, a)]
print(x)
print(a)
print(y)
print(np.exp(1)) # 자연상수의 1승
# 데이터 생성
data = {
    'Height': x,
    'Weight': y

}

# DataFrame 생성
df = pd.DataFrame(data)

# 산포도 그래프 그리기
df.plot(kind='scatter', x='Height', y='Weight', title='Height vs Weight')

# 그래프의 축 레이블 설정
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.grid()

# 그래프 보여주기
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
print(math.e) # 오일러 상수

p = -3 # 평행이동. 지수와 b에 값을 변화시켜도 그래프의 모양은 같다.
x = [i / 10 + p for i in range(-10,11)]
a = [2 for i in range(-10,11)] # 대칭이동. 지수나 밑에 -1 를 곱하면 대칭이 된다
y = [j ** i for i, j in zip(x, a)]
print(x)
print(a)
print(y)
print(np.exp(1)) # 자연상수의 1승
# 데이터 생성
data = {
    'Height': x,
    'Weight': y

}

# DataFrame 생성
df = pd.DataFrame(data)

# 산포도 그래프 그리기
df.plot(kind='scatter', x='Height', y='Weight', title='Height vs Weight')

# 그래프의 축 레이블 설정
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.grid()

# 그래프 보여주기
plt.show()
```

```python
# 로지스틱 logistic : 0 과 1 사이의 값을 갖는다.  f(x) = 1 / 1 + e ** -x
# sigmoid 그래프가 s 처럼 생겼음.  
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math

x = [i/100 for i in range(-600, 601)]
print(x)
y = []
for i, v in enumerate(x):
    y.append( 1 / (1 + (np.pow(np.e, -v))))
print(y)
data = {
    'Height':x,
    'Weight': y

}

# DataFrame 생성
df = pd.DataFrame(data)

# 산포도 그래프 그리기
df.plot(kind='scatter', x='Height', y='Weight', title='Height vs Weight')

# 그래프의 축 레이블 설정
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.grid()

# 그래프 보여주기
plt.show()
```
