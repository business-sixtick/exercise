# matplot

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = np.array([2,4,6,8])
x2 = np.array([0,4,2,3])
y = np.array([81,93,91,97])

plt.rcParams['font.family'] = 'gulim'
# plt.rcParams['fo']
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter3D(x1, x2, y)
ax.set_xlabel('')
ax.set_ylabel('')
ax.set_zlabel('')
print(ax)
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)

fig0 = plt.figure()
axes0 = fig0.add_axes([0,0,1,1]) # 
axes1 = fig0.add_axes([1,0,0,1])
axes0.plot(x, y)
axes1.plot(x, z)
plt.show()
```
