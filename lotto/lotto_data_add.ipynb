```python
import sqlite3
con = sqlite3.connect('../data/lotto.db')
cur = con.cursor()
# 테이블 목록 조회
cur.execute("SELECT name FROM sqlite_master WHERE type='table';") # sqlite_master 라는 테이블에 모든 스키마 정보를 저장하고 있음.
tables = cur.fetchall()

print("테이블 목록:")
for table in tables:
    print(table[0])

# 데이터베이스 연결 해제
con.close()
```

```python
import sqlite3
con = sqlite3.connect('../data/lotto.db')
cur = con.cursor()
query = '''create table if not exists win (
    "id" integer primary key,
    "win1" integer not null,
    "win2" integer not null,
    "win3" integer not null,
    "win4" integer not null,
    "win5" integer not null,
    "win6" integer not null,
    "bonus" integer not null)'''
cur.execute(query)
con.close()
```

```python
import openpyxl
import re # 정규식
import sqlite3
con = sqlite3.connect('../data/lotto.db')
cur = con.cursor()
wb = openpyxl.load_workbook(filename='../data/lotto_1_600.xlsx') #C:\source\exercise\data\lotto_1_600.xlsx
print(wb.sheetnames)
ws = wb['excel1_600']

# for row in ws.rows :
#     for cell in row :
#         row_values = []
#         if str(cell.value).isnumeric() :
#             row_values.append(cell.value)
#         print(row_values)
values = []

pattern = r'^\d{4}\.\d{2}\.\d{2}$'
count = 0
for row in ws.rows :
    values = [cell.value for cell in row]
    win = []
    if str(values[1]).isdigit() and re.match(pattern, values[2]) :
        print(values)
        print('count : ',count + 1 , 'id : ', values[1], 
              'win1 : ', values[-7],
               'win2 : ', values[-6],
                'win3 : ', values[-5],
                 'win4 : ', values[-4],
                  'win5 : ', values[-3],
                   'win6 : ', values[-2],
                    'bonus : ', values[-1], )
        count += 1
        win.append(values[1])
        win.append(values[-7])
        win.append(values[-6])
        win.append(values[-5])
        win.append(values[-4])
        win.append(values[-3])
        win.append(values[-2])
        win.append(values[-1])
        cur.execute('''INSERT INTO win (id, win1, win2, win3, win4, win5, win6, bonus) VALUES (?, ?, ?, ?, ?, ?, ?, ?)''', win)
    
con.commit()
con.close()
    
```
