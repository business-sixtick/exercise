```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns
import sixtick.python.easy as s
# 한글 폰트 경로 설정 (예: 나눔고딕)
font_path = r"C:\Users\hungh\AppData\Local\Microsoft\Windows\Fonts\NanumSquareRoundR.ttf"  # ttf 파일 경로

# 폰트 설정
fontprop = fm.FontProperties(fname=font_path, size=11)
print(fontprop)
plt.rcParams['font.family'] = fontprop.get_name()

# 마이너스 폰트 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False

# seaborn 스타일 설정
sns.set_theme(style="whitegrid")

# 예제 그래프
data = [1, 2, 3, 4, 5]
sns.lineplot(x=[0, 1, 2, 3, 4], y=data)

# 그래프 제목 설정 (한글로)
plt.title('한글 폰트 테스트', fontproperties=fontprop)
# plt.title('한글 폰트 테스트')

plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns
# 한글 폰트 설정 (예: 나눔고딕)
# font_path = r"C:\Users\hungh\AppData\Local\Microsoft\Windows\Fonts\NanumGothic.ttf"  # 실제 폰트 경로 입력
font_path = r"C:\Users\hungh\AppData\Local\Microsoft\Windows\Fonts\NanumSquareRoundR.ttf"  # 실제 폰트 경로 입력
fontprop = fm.FontProperties(fname=font_path)
plt.rcParams['font.family'] = fontprop.get_name()
plt.rcParams['axes.unicode_minus'] = False
print(fontprop.get_name())
# 예제 그래프
plt.figure(figsize=(8, 6))
plt.title('한글 폰트 설정 예제', fontproperties=fontprop)
# plt.plot([1, 2, 3], [4, 3, 5])
sns.lineplot(x=[1, 2, 3], y=[4, 3, 5])
plt.xlabel('X축', fontsize=14, fontproperties=fontprop)
plt.ylabel('Y축', fontsize=14, fontproperties=fontprop)
plt.show()
```
