```python
import cv2 as cv
import numpy as np
from PIL import Image, ImageDraw, ImageFont

# 이미지 크기
width, height = 500, 300

# 투명 배경의 빈 이미지 생성 (RGBA 모드)
image = Image.new('RGBA', (width, height), (255, 255, 255, 0))

# 폰트 설정
font_size = 60  # 폰트 사이즈를 크게 설정
font_path = r'C:\Users\hungh\AppData\Local\Microsoft\Windows\Fonts\NanumSquareRoundR.ttf' # 한글 폰트 파일 경로
font = ImageFont.truetype(font_path, font_size)

# 텍스트 그리기
draw = ImageDraw.Draw(image)
text = "안녕하세요, OpenCV와 Pillow!"
text_position = (50, 100)
text_color = (0, 0, 0, 255)  # 검정색 텍스트, 완전 불투명

# 텍스트의 bounding box 계산
bbox = draw.textbbox(text_position, text, font=font)

# 텍스트의 bounding box 좌표 추출
left, top, right, bottom = bbox

# 배경색 설정
background_color = (255, 255, 255, 255)  # 흰색 배경

# 배경 추가 (좌상단과 우하단 좌표를 리스트로 전달)
draw.rectangle([left, top, right, bottom], fill=background_color)

# 텍스트 그리기
draw.text(text_position, text, font=font, fill=text_color)

# 이미지를 OpenCV 형식으로 변환 (배경이 투명한 부분은 검은색으로 처리됨)
opencv_image = np.array(image)

# BGR 형식으로 변환 (OpenCV는 BGR 형식을 사용함)
opencv_image_bgr = cv.cvtColor(opencv_image, cv.COLOR_RGBA2BGR)

# 그레이스케일 변환
gray = cv.cvtColor(opencv_image_bgr, cv.COLOR_BGR2GRAY)

# 엣지 감지
edges = cv.Canny(gray, 50, 150)

# 허프 변환을 통한 직선 검출
lines = cv.HoughLinesP(edges, 1, np.pi / 180, 50, minLineLength=100, maxLineGap=10)

# 이미지 복사
line_image = np.copy(opencv_image_bgr)

# 검출된 선을 이미지에 그리기
if lines is not None:
    for line in lines:
        x1, y1, x2, y2 = line[0]
        cv.line(line_image, (x1, y1), (x2, y2), (0, 255, 0), 2)  # 초록색 선

# 결과 이미지 표시
cv.imshow('Detected Lines', line_image)
cv.waitKey(0)
cv.destroyAllWindows()
```
