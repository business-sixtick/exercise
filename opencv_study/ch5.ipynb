```python
import cv2
import numpy as np

# 이미지 읽기
img = cv2.imread('screenshot.png')

# HSV 색상 공간으로 변환
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
# Hue 채널 추출
hue_channel = hsv_img[:, :, 0]

# 특정 Hue 범위 설정 (예: 25에서 35 사이)
lower_bound = 28
upper_bound = 32

# Hue 범위 내의 픽셀을 30으로 설정, 범위 외의 픽셀을 120으로 설정
hue_channel_new = np.where((hue_channel > lower_bound) & (hue_channel < upper_bound), 30, 120)

# 변경된 Hue 채널을 HSV 이미지에 적용
hsv_img[:, :, 0] = hue_channel_new

# HSV 이미지를 BGR로 변환
bgr_img = cv2.cvtColor(hsv_img, cv2.COLOR_HSV2BGR)
bgr_img[:,:,0] = 0
# bgr_img = cv2.cvtColor(hsv_img, cv2.COLOR_BGR2GRAY)
ret, img_bin = cv2.threshold(bgr_img, 190, 255, cv2.THRESH_BINARY)
# 결과 출력
cv2.imshow('Filtered Image', img_bin)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# 이미지 읽기
image = cv2.imread('screenshot.png', cv2.IMREAD_GRAYSCALE)
# 이미지 반전
inverted_image = 255 - image
# Adaptive Mean Thresholding
mean_threshold = cv2.adaptiveThreshold(inverted_image, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 7, 2)

# Adaptive Gaussian Thresholding
gaussian_threshold = cv2.adaptiveThreshold(inverted_image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 3, 0)

# 결과 저장
# cv2.imwrite('mean_threshold.jpg', mean_threshold)
# cv2.imwrite('gaussian_threshold.jpg', gaussian_threshold)

# 결과 표시
cv2.imshow('Mean Threshold', mean_threshold)
cv2.imshow('Gaussian Threshold', gaussian_threshold)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python

        
import cv2 as cv
from win32api import GetSystemMetrics
screen_width = GetSystemMetrics(0)
screen_height = GetSystemMetrics(1)
# image_height, image_width = image.shape[:2]

img = cv.imread('다운로드.png')
image_name = 'dk'
cv.imshow(image_name, img)
image_height, image_width = img.shape[:2]
x, y = (screen_width // 2) - image_width // 2, screen_height // 2 - image_height // 2
cv.moveWindow(image_name, x, y) 
color = (0,0,0)
def onMouse(event, x, y, flag, param):
    
    if event == cv.EVENT_LBUTTONDOWN:
        cv.circle(img, (x,y), 50, color, 5) # -1 채움 , 기본적으로 선 굵기
        cv.imshow(image_name, img)
        print(event, x, y)
    elif event == cv.EVENT_RBUTTONDOWN:
        cv.putText(img, 'cho young hoon', (x, y), cv.FONT_HERSHEY_PLAIN, fontScale=5, color=color, thickness=5) 
        cv.imshow(image_name, img)
        print(event, x, y)

cv.setMouseCallback(image_name, onMouse)
while True:

    key = cv.waitKey(0)
    if key == ord('q'):
        break
    elif key == ord('h'):
        pass
    elif key == ord('1'):
        color = (0,0,0)
    elif key == ord('2'):
        color = (255,0,0)
    elif key == ord('3'):
        color = (0,255,0)
    elif key == ord('4'):
        color = (0,0,255)
    elif key == ord('5'):
        color = (0,255,255)
cv.destroyAllWindows() # 모든 화면 닫기
```
